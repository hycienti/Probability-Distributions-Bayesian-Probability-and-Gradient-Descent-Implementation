```python
import pandas as pd
df = pd.read_csv('IMDB Dataset.csv')
print(df.head())
positive_count = len(df[df['sentiment'] == 'positive'])
total_reviews = len(df)
p_positive = positive_count / total_reviews
print(f"Total Positive Reviews: {positive_count}")
print(f"Total Reviews: {total_reviews}")
print(f"P(Positive) = {p_positive:.4f}")
```

```python
keyword = "masterpiece"
positive_reviews = df[df['sentiment'] == 'positive']['review'].str.lower()
total_words_positive = positive_reviews.str.split().str.len().sum()
keyword_count_positive = positive_reviews.str.contains(rf'\b{keyword}\b').sum()
vocabulary_size = 1000
p_likelihood = (keyword_count_positive + 1) / (total_words_positive + vocabulary_size)
total_keyword_count = df['review'].str.lower().str.contains(rf'\b{keyword}\b').sum()
total_words_all = df['review'].str.lower().str.split().str.len().sum()
p_evidence = total_keyword_count / total_words_all
print(f"Likelihood P({keyword}|Positive): {p_likelihood:.6f}")
print(f"Evidence P({keyword}): {p_evidence:.6f}")
```

```python
p_posterior = (p_likelihood * p_positive) / p_evidence
print(f"--- Results for '{keyword}' ---")
print(f"Prior P(Positive): {p_positive:.4f}")
print(f"Likelihood P({keyword}|Positive): {p_likelihood:.6f}")
print(f"Evidence P({keyword}): {p_evidence:.6f}")
print(f"POSTERIOR P(Positive|{keyword}): {p_posterior:.4f}")
```
